```python
#필요한 라이브러리 불러오기
from random import *
```

```python
n=int(input("몇 개의 원소를 가진 배열을 만드시겠습니까?"))
```

```python
X=[] #초기 빈 배열 만들어주기
```

```python
for i in range(0,n): #O에서부터 n-1까지 배열 생성
  #X.append(i+1) #숫자를 차례대로 입력
  #X.append(randint(1,n)) #랜덤하게 입력
  X.append(randint(1,n*2))
print(X)
```

```python
A=list(range(0,n))
```

```python
print(A)
```

```python
for i in range(0,n):
  s=X[0]
  for j in range(1,i+1):
    s=s+X[j]
  A[i]=s/(i+1)
```
